```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import  load_model
import streamlit as st
import numpy as np 

st.header('Image Classification Model')
model = load_model('Image_classify.keras')
data_cat = ['Orange_orange',
 'black_purple_grapes',
 'brown_banana',
 'green karela',
 'green_apple',
 'green_banana',
 'green_capsicum',
 'green_grapes',
 'green_yelloe_orange',
 'orange_capsicum',
 'red_apple',
 'red_capsicum',
 'red_grapes',
 'red_yellow_apple',
 'yellow karela',
 'yellow_banana']
img_height = 180
img_width = 180
image =st.text_input('Enter Image name','corn.jpg')

image_load = tf.keras.utils.load_img(image, target_size=(img_height,img_width))
img_arr = tf.keras.utils.array_to_img(image_load)
img_bat=tf.expand_dims(img_arr,0)

predict = model.predict(img_bat)

score = tf.nn.softmax(predict)
st.image(image, width=200)
st.write('Veg/Fruit in image is ' + data_cat[np.argmax(score)])
st.write('With accuracy of ' + str(np.max(score)*100))
```

```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import numpy as np
from PIL import Image
import os

# Load the model and add a diagnostic print
try:
    model = load_model('Image_classify.keras')
    print("Model loaded successfully.")
except Exception as e:
    print(f"Error loading model: {e}")

# Define categories
data_cat = ['Orange_orange',
 'black_purple_grapes',
 'brown_banana',
 'green karela',
 'green_apple',
 'green_banana',
 'green_capsicum',
 'green_grapes',
 'green_yelloe_orange',
 'orange_capsicum',
 'red_apple',
 'red_capsicum',
 'red_grapes',
 'red_yellow_apple',
 'yellow karela',
 'yellow_banana']

img_height = 180
img_width = 180
image_path = os.path.join('corn.jpg')  # Use your test image name here

if not os.path.exists(image_path):
    print(f"Image not found: {image_path}")
    exit()

# Load and preprocess the image
try:
    print(f"Selected image: {image_path}")
    image_load = tf.keras.utils.load_img(image_path, target_size=(img_height, img_width))
    img_arr = tf.keras.preprocessing.image.img_to_array(image_load)
    img_bat = np.expand_dims(img_arr, axis=0)

    # Make prediction
    print("Making prediction...")
    predict = model.predict(img_bat)
    print(f"Raw prediction output: {predict}")
    
    # Apply softmax to get probabilities
    score = tf.nn.softmax(predict)
    print(f"Softmax probabilities: {score}")

    # Print the prediction results
    predicted_class = data_cat[np.argmax(score)]
    confidence = np.max(score) * 100
    print(f'Predicted class: {predicted_class} with accuracy: {confidence:.2f}%')

except Exception as e:
    print(f"Error processing image: {e}")
```

```python
import serial
import time
import threading

esp32_port = 'COM9'  # Update with your actual COM port
baud_rate = 115200

# Variables to store temperature and humidity
T = None
H = None
stop_thread = False  # Control variable to stop the thread

def read_serial_data():
    global T, H, stop_thread
    try:
        ser = serial.Serial(esp32_port, baud_rate, timeout=2)  # Increase timeout to avoid issues
        time.sleep(2)  # Allow some time for the connection to establish

        while not stop_thread:  # Keep reading while the thread is active
            data = ser.readline().decode('utf-8').strip()
            if data:  # Only process if data is not empty
                print(data)
                # Assuming data is in the format: "Temperature: 30.50 °C, Humidity: 70.80 %"
                try:
                    # Split the data string to extract temperature and humidity values
                    temp_str = data.split(",")[0].split(":")[1].strip().split(" ")[0]
                    hum_str = data.split(",")[1].split(":")[1].strip().split(" ")[0]

                    # Convert strings to integers
                    T = float(temp_str)
                    H = float(hum_str)

                    # Print the extracted integer values
                    # print(f"Temperature (T): {T} °C, Humidity (H): {H} %")
                except (IndexError, ValueError) as parse_error:
                    print(f"Failed to parse data: {data}. Error: {parse_error}")

    except Exception as e:
        print(f"Error: {e}")

    finally:
        ser.close()
        print("Closed serial port")

# Start the thread for reading serial data
serial_thread = threading.Thread(target=read_serial_data)
serial_thread.start()

# You can stop the thread when needed by setting stop_thread to True
# Example: stop_thread = True, then call serial_thread.join() to wait for it to finish
```
